```python
import os
import cv2
import torch
import numpy as np
from pathlib import Path
import torchvision
from diffusers import StableDiffusionPipeline
```

# Image Generation

```python
def generate_image(prompt: str, output_path: Path) -> None:
    model_id = "CompVis/stable-diffusion-v1-4"
    pipe = StableDiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16
    )
    pipe = pipe.to("cuda" if torch.cuda.is_available() else "cpu")

    image = pipe(prompt).images[0]
    image.save(str(output_path))
    return
```

```python
prompts = [
    "A single red apple on a white background.",
    "A blue cup placed on a simple wooden table.",
    "A country side scene with a barn, bike and a few cattle",
    "Multiple people walking on a busy market street",
    "A classic car parked on a quiet street with a  backdrop.",
    "A neatly arranged bowl of fruit on a  countertop.",
    "A black and white cat sitting on a sofa.",
    "A modern chair in a room with a plain, neutral background.",
    "A single coffee mug on a solid-colored table.",
    "A vase with one flower on a  desk."
]
```

```python
output_dir = "generated_images"
os.makedirs(output_dir, exist_ok=True)
```

```python
from diffusers import DiffusionPipeline
for i, prompt in enumerate(prompts, start=1):
        output_path = os.path.join(output_dir, f"image_{i:02d}.png")
        print(f"Generating image for prompt {i}: {prompt}")
        generate_image(prompt, output_path)
```

# Object Detection

```python
# Load YOLOv5 via PyTorch Hub
print("Loading YOLOv5 model...")
yolo_model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
yolo_model.conf = 0.5  # set confidence threshold

# Setup Detectron2 Predictor
print("Setting up Detectron2 model...")
import detectron2
from detectron2.engine import DefaultPredictor
from detectron2.config import get_cfg
from detectron2 import model_zoo
from detectron2.data import MetadataCatalog

cfg = get_cfg()
cfg.merge_from_file(model_zoo.get_config_file("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml"))
cfg.MODEL.ROI_HEADS.SCORE_THRESH_TEST = 0.5  # set threshold for detection
cfg.MODEL.WEIGHTS = model_zoo.get_checkpoint_url("COCO-InstanceSegmentation/mask_rcnn_R_50_FPN_3x.yaml")
detectron2_predictor = DefaultPredictor(cfg)
detectron2_metadata = MetadataCatalog.get(cfg.DATASETS.TRAIN[0])

# YOLOv5 Detection Function

def detect_yolov5(image: np.ndarray) -> list:
    """
    Run YOLOv5 detection on the image.
    Returns a list of detection dictionaries with keys: xmin, ymin, xmax, ymax, confidence, name.
    """
    results = yolo_model(image)
    detections = results.pandas().xyxy[0].to_dict(orient="records")
    return detections

# Detectron2 Detection Function

def detect_detectron2(image: np.ndarray) -> list:

    outputs = detectron2_predictor(image)
    instances = outputs["instances"].to("cpu")
    boxes = instances.pred_boxes.tensor.numpy()  # Nx4 array
    scores = instances.scores.numpy()  # N array
    classes = instances.pred_classes.numpy()  # N array

    detections = []
    for box, score, cls in zip(boxes, scores, classes):
        xmin, ymin, xmax, ymax = box
        label = detectron2_metadata.thing_classes[cls]
        detections.append({
            "xmin": float(xmin),
            "ymin": float(ymin),
            "xmax": float(xmax),
            "ymax": float(ymax),
            "confidence": float(score),
            "name": label
        })
    return detections

# Ensemble and NMS Function
def ensemble_detections(dets1: list, dets2: list, iou_threshold: float = 0.5) -> list:
    # Combine the two lists
    combined = dets1 + dets2

    # Group detections by label (class name)
    grouped = {}
    for det in combined:
        label = det["name"]
        grouped.setdefault(label, []).append(det)

    final_detections = []
    for label, detections in grouped.items():
        boxes = []
        scores = []
        for det in detections:
            boxes.append([det["xmin"], det["ymin"], det["xmax"], det["ymax"]])
            scores.append(det["confidence"])
        boxes_tensor = torch.tensor(boxes)
        scores_tensor = torch.tensor(scores)
        # Run NMS from torchvision
        keep_indices = torchvision.ops.nms(boxes_tensor, scores_tensor, iou_threshold)
        for idx in keep_indices:
            final_detections.append(detections[idx])
    return final_detections

# Visualization Function
def visualize_detections(image: np.ndarray, detections: list) -> np.ndarray:
    vis_image = image.copy()
    for det in detections:
        xmin = int(det["xmin"])
        ymin = int(det["ymin"])
        xmax = int(det["xmax"])
        ymax = int(det["ymax"])
        label = det["name"]
        score = det["confidence"]
        cv2.rectangle(vis_image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
        cv2.putText(vis_image, f"{label}: {score:.2f}", (xmin, max(ymin - 10, 10)),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    return vis_image
```

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Main Ensemble Pipeline
def run_ensemble_pipeline(input_folder: str = "generated_images", output_folder: str = "annotated_detections"):
    input_path = Path(input_folder)
    output_path = Path(output_folder)
    output_path.mkdir(exist_ok=True)

    # Process each image in the input folder
    for img_file in input_path.glob("*.png"):
        print(f"Processing {img_file.name}...")
        image = cv2.imread(str(img_file))
        if image is None:
            print(f"Failed to load image {img_file.name}. Skipping.")
            continue

        # Run both detectors
        dets_yolo = detect_yolov5(image)
        dets_detectron2 = detect_detectron2(image)

        # Combine detections using ensemble method (with NMS)
        ensemble = ensemble_detections(dets_yolo, dets_detectron2, iou_threshold=0.5)

        # Visualize the combined detections
        vis_image = visualize_detections(image, ensemble)

        # Save the output image
        out_file = output_path / img_file.name
        cv2.imwrite(str(out_file), vis_image)
        print(f"Saved ensemble detection output to {out_file}")

if __name__ == "__main__":
    run_ensemble_pipeline()
```
